```python
import pandas as pd
import numpy as np


batters_data = pd.read_csv(r"C:\Users\ASUS\OneDrive\Documents\IPL Data Analysis\datasets\fact_bating_summary.csv")
match_data = pd.read_csv(r"C:\Users\ASUS\OneDrive\Documents\IPL Data Analysis\datasets\dim_match_summary.csv")
batters_data = pd.merge(batters_data, match_data, on="match_id", how="left")
batters_data["matchDate"] = pd.to_datetime(batters_data["matchDate"])
batters_data['year'] = batters_data["matchDate"].dt.year
# 1. Top 10 batsmen based on past 3 years total runs scored
top_10_batters = batters_data.groupby("batsmanName").sum("runs").sort_values('runs', ascending=False).head(10).reset_index()
top_10_batters = top_10_batters[["batsmanName", "runs"]]
top_10_batters.to_excel(r'./Output.xlsx', sheet_name= 'Top_10_batters(score)')
```

```python

# Top 10 batsmen based on past 3 years batting average. (min 60 balls faced in each season)
batters_out_data = batters_data.loc[batters_data["out/not_out"] == 'out'].groupby("batsmanName").count().reset_index()[['batsmanName','out/not_out']]
no_of_60balls_season = batters_data.groupby(['year','batsmanName']).sum('balls').loc[batters_data.groupby(['year','batsmanName']).sum('balls')['balls'] > 60].reset_index()[['year','batsmanName','balls']]
filtered = batters_data[batters_data['batsmanName'].isin(no_of_60balls_season['batsmanName'])]
main_df= pd.merge(filtered,batters_out_data,on='batsmanName' , how='left').sort_values(by='batsmanName', ascending=True)
filtered_df = filtered[['batsmanName','runs','balls']].groupby('batsmanName').sum('runs').reset_index().sort_values(by='batsmanName', ascending=True)
out_df = pd.merge(filtered_df,batters_out_data,on='batsmanName' , how='left').sort_values(by='batsmanName', ascending=True)
out_df['batting_avg'] = round(out_df['runs'] / out_df['out/not_out'],2)
out_df.sort_values(by = 'batting_avg', inplace=True, ascending=False)
top_10_batters_avg = out_df[['batsmanName','runs','balls','batting_avg']].head(10)
top_10_batters_avg.to_excel(r'./Output.xlsx', sheet_name= 'Top_10_batters(Avg)')

top_10_batters_avg
# out_df
# out_df.to_excel(r'./Output.xlsx', sheet_name= 'Top_10_batters(Avg)')
```

```python
# Top 10 batsmen based on past 3 years strike rate (min 60 balls faced in each season)
filtered = batters_data[batters_data['batsmanName'].isin(no_of_60balls_season['batsmanName'])]
strikeData = filtered.groupby("batsmanName").sum("runs").reset_index()
strikeData['strikeRate'] = round(((strikeData['runs'] / strikeData['balls']) * 100),2)
Top_10_SR_batsmen = strikeData[['batsmanName','runs','balls','strikeRate']].sort_values(by='strikeRate', ascending=False).head(10)
Top_10_SR_batsmen
Top_10_SR_batsmen.to_excel(r'./Output.xlsx', sheet_name= 'Top_10_batters(SR)')
```

```python
# Top 5 batsmen based on past 3 years boundary % (fours and sixes).
filtered['boundary'] = filtered['4s'] + filtered['6s']
batters_data1 = filtered.groupby('batsmanName').sum('boundary').reset_index()
batters_data1['boundary%'] = round((batters_data1['boundary'] / batters_data1['balls'])*100,2)
batters_data['boundary%'] = round((batters_data1['boundary'] / batters_data1['balls'])*100,2)
Top_5_boundary_batsmen = batters_data1[['batsmanName', 'runs','balls' ,'4s','6s','boundary', 'boundary%']].sort_values(by='boundary%', ascending=False).fillna(0.00).head(5)
Top_5_boundary_batsmen.to_excel(r'./Output.xlsx', sheet_name= 'Top_10_batters(boundary%)')
Top_5_boundary_batsmen
```

```python
# Top 10 bowlers based on past 3 years total wickets taken.
bowlers_data = pd.read_csv(r"C:\Users\ASUS\OneDrive\Documents\IPL Data Analysis\datasets\fact_bowling_summary.csv")
match_data = pd.read_csv(r"C:\Users\ASUS\OneDrive\Documents\IPL Data Analysis\datasets\dim_match_summary.csv")
bowlers_data = pd.merge(bowlers_data, match_data, on="match_id", how="left")
bowlers_data["matchDate"] = pd.to_datetime(bowlers_data["matchDate"])
bowlers_data['year'] = bowlers_data["matchDate"].dt.year
top_10_bowlers = bowlers_data.groupby('bowlerName').sum('wickets').reset_index().sort_values(by='wickets', ascending=False).head(10)[['bowlerName','wickets']]
top_10_bowlers
top_10_bowlers.to_excel(r'./Output.xlsx', sheet_name= 'top_10_bowlers(wickets)')
```

```python
# Top 10 bowlers based on past 3 years bowling average. (min 60 balls bowled in each season)
bowlers_df = bowlers_data.groupby(['year','bowlerName']).sum('overs').reset_index().groupby('bowlerName').sum('overs').reset_index()
bowlers_df['balls'] =  bowlers_df['overs'] * 6
bowlers_df = bowlers_df[bowlers_df['balls'] > 60]
bowlers_df['average'] = round(bowlers_df['runs']/bowlers_df['wickets'],2)
top_10_bowlers_avg = bowlers_df[['bowlerName','runs','wickets','economy','balls','average']].sort_values(by='average', ascending=True).head(10)
top_10_bowlers_avg.to_excel(r'./Output.xlsx', sheet_name= 'top_10_bowlers(Avg)')
top_10_bowlers_avg
```

```python
# Top 10 bowlers based on past 3 years economy rate. (min 60 balls bowled in each season)
bowlers_df = bowlers_data.groupby(['year','bowlerName']).sum('overs').reset_index().groupby('bowlerName').sum('overs').reset_index()
bowlers_df['balls'] =  bowlers_df['overs'] * 6
bowlers_df = bowlers_df[bowlers_df['balls'] > 60]
bowlers_df['economy'] = round(bowlers_df['runs']/bowlers_df['overs'],2)
top_10_bowlers_economy = bowlers_df[['bowlerName','runs','balls','overs','wickets','economy']].sort_values(by='economy', ascending=True).head(10)
# top_10_bowlers_economy.to_excel(r'./Output.xlsx', sheet_name= 'top_10_bowlers(economy)')
top_10_bowlers_economy
```

```python
# Top 5 bowlers based on past 3 years dot ball %.
bowlers_df = bowlers_data.groupby(['year','bowlerName']).sum('overs').reset_index().groupby('bowlerName').sum('overs').reset_index()
bowlers_df['balls'] =  bowlers_df['overs'] * 6
bowlers_df = bowlers_df[bowlers_df['overs'] > 100]
bowlers_df['dot_avg'] = round(bowlers_df['0s'] / bowlers_df['balls'],2) *100
bowlers_df.sort_values(by='dot_avg', ascending=False, inplace=True)
top_5_bowlers_dot = bowlers_df[['bowlerName','runs','overs','wickets','economy','balls','dot_avg']].head(5)
top_5_bowlers_dot.to_excel(r'./Output.xlsx', sheet_name= 'top_5_bowlers(dot)')
top_5_bowlers_dot
```

```python
# Top 4 teams based on past 3 years winning %.
match_summary = pd.read_csv(r"C:\Users\ASUS\OneDrive\Documents\IPL Data Analysis\datasets\dim_match_summary.csv")
match_summary["matchDate"] = pd.to_datetime(match_summary["matchDate"])
match_summary['year'] = match_summary["matchDate"].dt.year
# match_summary.groupby(['winner','year']).count().reset_index()
all_teams = list(set(match_summary['team1']).union(set(match_summary['team2'])))
total_matches_played = {}
for team_name in all_teams:
   total_matches_played[team_name] = ((match_summary['team1'] == team_name).sum()) + ((match_summary['team2'] == team_name).sum())
Team = []
matches = []
for i,j in total_matches_played.items():
   Team.append(i)
   matches.append(j)

Match_desc = pd.DataFrame({'Team':Team,'Matches':matches})
match_summary = match_summary.groupby('winner').count().reset_index()[['winner','team1']].sort_values(by='winner', ascending=True).rename(columns={'winner':'Team','team1':'Winning_count'})
Match_desc.sort_values(by='Team', ascending=True,inplace=True)
top_teams = pd.merge(Match_desc, match_summary, on='Team', how='inner')
top_teams['winning_%'] = round(((top_teams['Winning_count'] / top_teams['Matches']) * 100),2)
top_teams.sort_values(by='winning_%', ascending=False, inplace=True)
top_4_teams = top_teams.head(4)
top_4_teams
top_4_teams.to_excel(r'./Output.xlsx', sheet_name= 'top_4_teams(winning%)')
```
